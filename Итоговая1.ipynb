```python
df_left = df[df['left']==1]
df_stay = df[df['left']==0]
promotion_share_left = round(len(df_left[ df_left['promotion_last_5years']> 0])/len(df_left)*100,2)
promotion_share_stay = round(len(df_stay[ df_stay['promotion_last_5years']> 0])/len(df_stay)*100,2)
```
